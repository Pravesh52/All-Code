#  Task1. Customer Personality Analysis

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("customer_personality.csv")

# 1. Rename columns
df.columns = df.columns.str.lower().str.replace(" ", "_")

# 2. Remove duplicates
df = df.drop_duplicates()

# 3. Fix education names
df['education'] = df['education'].replace({
    '2n Cycle': 'Secondary',
    'Basic': 'Basic Education',
    'Graduation': 'Graduate',
    'PhD': 'PhD'
})

# 4. Fix marital status
df['marital_status'] = df['marital_status'].replace({
    'Married': 'Married',
    'Single': 'Single',
    'Divorced': 'Divorced',
    'Widow': 'Widowed',
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd': 'Other'
})

# 5. Convert date formats
# 5. Convert date formats (FIXED)
df['dt_customer'] = (
    df['dt_customer']
      .astype(str)
      .str.replace('/', '-', regex=False)
      .str.replace('.', '-', regex=False)
)

df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True, errors='coerce')

# Fill failed dates (rare)
df['dt_customer'] = df['dt_customer'].fillna(df['dt_customer'].mode()[0])

```
