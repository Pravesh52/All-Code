1. Create Array from student dataset math scores


```python
math_scores_array=df['math_original'].to_numpy()
print("Math scores Array:")
print(math_scores_array)
```

Generate Grade boundaries using linspace for science scores 

```python
min_science=df['science_original'].min()
max_science=df['science_original'].max()

grade_boundaries=np.linspace(min_science,max_science,5)
print(grade_boundaries)

```

Initialize digital attendance system with zeros

```python
num_students=len(df)
attend=np.zeros(num_students,dtype=int)
```

4.Optimize memory usage for student performance grades


```python
print(f"original 'grade' dtype: {df['grade'].dtype}")
print("Before optimization memory usage:")
print(df['grade'].memory_usage(deep=True))
df['grade']=df['grade'].astype('category')
print(f"optimized 'grade' dtype:{df['grade'].dtype}")
print("Memory usage after optimization:")
print(df['grade'].memory_usage(deep=True))
```

Optimize storage for math scores using different integer types

```python
print(f"original 'math_original' dtype: {df['math_original'].dtype}")
print("Before optimization memory usage:")
print(df['math_original'].memory_usage(deep=True))

df['math_int8'] = df['math_original'].astype('int8')
print(f"\noptimized 'math_int8' dtype: {df['math_int8'].dtype}")
print("Memory usage after optimization:")
print(df['math_int8'].memory_usage(deep=True))

# convert attendance -> uint8 (correct)
```

```python
print(f"original 'attend' dtype: {attend.dtype}")
print("Before optimization memory usage:")
print(attend.nbytes)

attend_int8 = attend.astype('int8')
print(f"\noptimized 'attend_int8' dtype: {attend_int8.dtype}")
print("Memory usage after optimization:")
print(attend_int8.nbytes)
```

6.Explore float precision with science_original Scores

```python
print(f"original 'science_original' dtype: {df['science_original'].dtype}")
print("Before optimization memory usage:")
print(df['science_original'].memory_usage(deep=True))

df['science_float16'] = df['science_original'].astype('float16')
print(f"\noptimized 'science_float16' dtype: {df['science_float16'].dtype}")
print("Memory usage after optimization:")
print(df['science_float16'].memory_usage(deep=True))
```

Optimize Memory for attendance DATA


```python
print(f"original 'attend' dtype: {attend.dtype}")
print("Before optimization memory usage:")
print(attend.nbytes)

attend_uint8 = attend.astype('uint8')
print(f"\noptimized 'attend_uint8' dtype: {attend_uint8.dtype}")
print("Memory usage after optimization:")

print(attend_uint8.nbytes)
```
