Python Program to check where a string is palindrome or not?

```python
str="Pravesh"
l=0
k=len(str)-1
is_palindrome=True
while(l<k):
    if str[l]!=str[k]:
        is_palindrome=False
        break
    l+=1
    k-=1

if (is_palindrome):
    print("palindrome")
else:
    print("Not Palindrome")
    





```

```python
# 2nd way
mystr="madam"
mystr=mystr.lower()
revstr=reversed(mystr)

if list(mystr)==list(revstr):
    print("plaindrome")
else:
    print("not palindrome")
```
